## Features

Different features about each of the feature in the data:

* id (Unique id for each patient)
* age (Age of the patient in years)
* origin (place of study)
* sex (Male/Female)
* cp chest pain type ([typical angina, atypical angina, non-anginal, asymptomatic])
* trestbps resting blood pressure (resting blood pressure (in mm Hg on admission to the hospital))
* chol (serum cholesterol in mg/dl)
* fbs (if fasting blood sugar > 120 mg/dl)
* restecg (resting electrocardiographic results) -- Values: [normal, stt abnormality, lv hypertrophy]
* thalach: maximum heart rate achieved
* exang: exercise-induced angina (True/ False)
* oldpeak: ST depression induced by exercise relative to rest
* slope: the slope of the peak exercise ST segment
* ca: number of major vessels (0-3) colored by fluoroscopy
* thal: [normal; fixed defect; reversible defect]
* num: the predicted attribute

```python
# changing sex to binary
df.loc[df['sex']=='Male', 'sex']=1
df.loc[df['sex']=='Female','sex']=0

# changing restecg to 0,1,2
def conditions(row):
    if row['restecg'] == 'normal':
        val = 0
    elif row['restecg'] == 'st-t abnormality':
        val = 1
    elif row['restecg'] == 'lv hypertrophy':
        val = 2  
    else:
        val = 3
    return val

#Apply the function to each data point in the data frame
df['restecg']= df.apply(conditions, axis=1)

# changing fbs to binary
df['fbs'] = np.where(df['fbs']== True, 1, 2)

# changing exang
df['exang'] = np.where(df['exang']== True, 1, 2)

# changing cp
def conditions(row):
    if row['cp'] == 'typical angina':
        val = 0
    elif row['cp'] == 'atypical angina':
        val = 1
    elif row['cp'] == 'non-anginal':
        val = 2  
    elif row['cp'] == 'asymptomatic':
        val = 3  
    else:
        val = 3
    return val

#Apply the function to each data point in the data frame
df['cp']= df.apply(conditions, axis=1)

# changing the slope
def conditions(row):
    if row['slope'] == 'flat':
        val = 0
    elif row['slope'] == 'downsloping':
        val = 1
    elif row['slope'] == 'upsloping':
        val = 2  
    else:
        val = 3
    return val

# Apply the function to each data point in the data frame
df['slope']= df.apply(conditions, axis=1)

# changing the thal
def conditions(row):
    if row['thal'] == 'normal':
        val = 0
    elif row['thal'] == 'reversable defect':
        val = 1
    elif row['thal'] == 'fixed defect':
        val = 2
    else:
        val = 3
    return val

#Apply the function to each data point in the data frame
df['thal']= df.apply(conditions, axis=1)
```

```python
df.head(10)
```

```python
x=df.drop(["target","dataset"],axis=1)
y=df["target"]
np.random.seed(42)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Models to use
models={"LR":LogisticRegression(),
       "KNN":KNeighborsClassifier(),
        "RC":RandomForestClassifier()}

# Function to use and measure the accuracy of models
def test_and_score(models,x_train,x_test,y_train,y_test):
    np.random.seed(42)
    model_score = {}
    for name, model in models.items():
        model.fit(x_train,y_train)
        model_score[name] = model.score(x_test, y_test)
    return model_score    
```

```python
results = test_and_score(models,x_train=x_train,x_test=x_test,y_train=y_train,y_test=y_test)
results
```

## Hyper_parameters Tuning


```python
train_set = []
test_set =[]

# Value of neighbors to run through
neighbors = range(1,21)

# Set up knn instance
knn = KNeighborsClassifier()

#Loop through different n_neighbors
for i in neighbors:
    knn.set_params(n_neighbors=i)
    knn.fit(x_train,y_train)
    train_set.append(knn.score(x_train,y_train))
    test_set.append(knn.score(x_test,y_test))
```

```python
train_set
```

```python
test_set
```

```python
plt.plot(neighbors, train_set, label="Train score")
plt.plot(neighbors, test_set, label="Test score")
plt.xticks(np.arange(1,21,1))
plt.xlabel("Number of neighbors")
plt.ylabel("Model Score")
plt.legend()

print(f"Maximum KNN score on the test data: {max(test_set)*100:.2f}%")
```

## Hyperparameter tuning with GridSearchCV

```python
# Hyperparameter grid for Logisctic Regression
log_reg_grid ={"C": np.logspace(-13,13,40),
              "solver": ["liblinear"]}

# Hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 100, 50),
          "max_depth": [None,3,5,10],
          "min_samples_split": np.arange(2,20,2),
          "min_samples_leaf": np.arange(1,20,2)}
```
